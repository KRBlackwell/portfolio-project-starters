## The code directly in the notebook.
You can copy and paste the code into your IDE or an editor or a notebook to run yourself.\
IPython's display SVG will help us display the result. However, the result appears "upside-down" compared with the above. Likely because of the different defaults and how it's creating and rendering the svg in the website.

```python
# !pip install pygal
import pygal
print("pygal version:", pygal.__version__)
from IPython.display import SVG, display
#I added "width, height, and explicit_size" to make the image smaller
#style=DarkStyle to match the trinket one and make it not so bright

# Our data is a list of series.
# Each series is a list with a label and a list of data points
data = [["Firefox", [None, None, 0, 16.6,   25,   31, 36.4, 45.5, 46.3, 42.8, 37.1]],
        ["Chrome",  [None, None, None, None, None, None,    0,  3.9, 10.8, 23.8, 35.3]],
        ["IE",      [85.8, 84.6, 84.7, 74.5,   66, 58.6, 54.7, 44.8, 36.2, 26.6, 20.1]],
        ["Others",  [14.2, 15.4, 15.3,  8.9,    9, 10.4,  8.9,  5.8,  6.7,  6.8,  7.5]]]


stackedline_chart = pygal.StackedLine(fill=True, width=500, height=250, explicit_size=True, style=pygal.style.DarkStyle)
stackedline_chart.title = "Browser usage evolution (in %)"
stackedline_chart.x_labels = map(str, range(2002, 2013))

for label, data_points in data:
    stackedline_chart.add(label, data_points)
```

```python
display(SVG(stackedline_chart.render()))
```
